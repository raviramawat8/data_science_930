# Decorators

```python
def decorate(func):
    def hello(*args):
        print("*"*100)
        print("-"*100)
        func(*args)
        print("-"*100)
        print("*"*100)
    return hello
```

```python
@decorate
def add(a,b):
    print(f"Sum of {a} and {b} is {a+b}".center(100))
```

```python
add(44,5)
```

```python
@decorate
def calc(a,b,op):
    if op == "+":
        print(f"Sum of {a} and {b} is {a+b}".center(100))
    elif op == "-":
        print(f"Sub of {a} and {b} is {a-b}".center(100))
    elif op == "*":
        print(f"Mul of {a} and {b} is {a*b}".center(100))
    else:
        print("Invalid operator")
```

```python
calc(33,3,'*')
```

```python
import time
import psutil
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib notebook
```

```python
plt.rcParams['animation.html']='jshtml'
fig = plt.figure(figsize=(10,4))
a1 = fig.add_subplot(111)
fig.show()
```

```python
i=0
x,y=[],[]
while True:
    x.append(i)
    y.append(psutil.cpu_percent())
    a1.plot(x,y)
    fig.canvas.draw()
    a1.set_xlim(left=max(0,i-50),right=i+50)
    time.sleep(1)
    i+=1
```
