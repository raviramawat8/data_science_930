```python
n = int(input())
c = 0
a = 0
b = 1
while c<n: 
    a,b = a+b,a
    print(a,end="\t")
    c+=1
```

```python
n = int(input())
d = 0
a = 0
b = 1
while d<n: 
    c = a+b
    b=a
    a=c
    print(a,end="\t")
    d+=1
```

```python
from math import sqrt
num = int(input())
c = 2
check = round(sqrt(num)+1)
while c<check:
    if num%c==0:
        print("Not a prime no.")
        break
    c+=1
else:
    print("Prime No.")
```
