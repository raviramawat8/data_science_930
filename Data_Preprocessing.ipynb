```python
gener = []
for var in range(len(data)-1):
    gener.extend(data[var].split('::')[2].split("|"))
```

```python
gener= list(set(gener))
```

```python
for g in gener:
    print(g)
```

```python
gener=sorted(gener)
```

```python
movies = {}
for g in gener:
    movies.setdefault(g,dict())
```

```python
movies
```

```python
year=[]
for line in data:
    if line:
        y = line.split("::")[1][-5:-1]
        if y not in year:
            year.append(y)
```

```python
year = sorted(year)
```

```python
year
```

```python
for key,value in movies.items():
    for y in year:
        value.setdefault(y,[])
```

```python
movies
```

```python
for line in data:
    if line:
        movie_name = line.split("::")[1][:-6]
        movie_year = line.split("::")[1][-5:-1]
        geners = line.split("::")[-1].split("|")
        for g in geners:
            movies[g][movie_year].append(movie_name)
```
